## **함수란 무엇인가요?**
**📌 함수란?**
* 함수는 입력값을 받아 처리하고, 결과를 반환하는 코드의 덩어리입니다.   


**📌 파이썬의 함수와 수학의 함수**
* 파이썬의 함수와 수학의 함수는 목적과 구현 방식에서 차이가 있지만, 모두 입력값을 받아 일정한 수식으로 처리한 결과를 반환한다는 공통점이 있습니다.
* 수학에서 함수를 공부할 때 배웠던 일차함수를 생각해봅시다. 
> 예를 들어 `f(x) = 7x+3`라는 함수의 x에 3을 넣으면, 결과 7*3+3=24를 반환해줍니다.
* 즉 입력에 대해 일정한 수식을 적용하여 결과를 반환합니다.  
* 이와 같이, 컴퓨터의 함수도 입력값에 대해 정의된 기능을 처리하는 일을 합니다.   





**📌 함수의 필요성**
* 함수는 재사용이 가능한 코드로, 코드를 재사용하고 중복을 제거할 수 있습니다. 
* 프로그램의 구조를 단순하게 만들어줍니다. 
* 코드의 가독성을 높이고 유지보수성을 개선시킵니다.

## **함수 정의, 호출 방법**





#### **함수 정의** 
* def : def 는 'define'의 약자로 '정의하다'라는 뜻입니다. 즉 함수를 '정의'한다는 의미로 가장 앞쪽에 def를 써준 것입니다. 
* 함수명 : 함수명은 변수명과 마찬가지로 숫자, 알파벳 등을 사용해서 만들 수 있습니다.  
* 매개변수 : 함수를 호출할 때 넣어준 인수가 함수 내부에서 활용될 수 있도록 해주는 변수입니다. 
* 명령문 : 함수를 호출하여 실행하고자 하는 명령문을 작성합니다. 명령문은 들여쓰기를 하고 작성합니다.    
* return : 함수의 반환값을 반환함과 동시에 함수를 종료합니다. 

#### **함수 호출** 
* 함수 정의 당시 매개변수를 생략했다면, 함수를 호출할 때에도 인수를 생략한 채 빈 괄호로 호출합니다. 
* 함수 정의 당시의 매개변수의 개수만큼 인수를 작성해주어야 합니다. 

#### **예제 1 : 함수 활용하기 - 인수가 없는 경우**

```python
#예제_1_함수_활용하기_인수가_없는_경우_

# 함수를 정의하는 부분입니다. 
def codle(): 
    print("코들")
    
# 함수를 호출하는 부분입니다. 
codle()
```

#### 예제 1 해설
> codle이라는 함수는 "코들"를 출력해주는 함수입니다.   
> 함수를 정의할 때 콜론(:)을 빠트리면 오류가 발생합니다.  
>
> 함수를 정의할 때는 코드가 실행되지 않고, 함수를 호출해야만 함수 내부 코드가 실행됩니다.   
> 함수를 정의만 하고 호출하지 않을 경우, 아무것도 출력되지 않을 것입니다. 

#### **예제 2 : 함수 활용하기 - 인수가 있는 경우**

```python
#예제_2_함수_활용하기_인수가_있는-경우
def calculation(x):
    y = 7*x +3
    return y  # 결과값 y를 반환합니다. 

n = calculation(7)  # 변수 n에 함수에서 반환된 y값이 저장됩니다. 
print("계산한 값은? "+ str(n))
```

#### 예제 2 해설
> calculation이라는 함수는 전달 받은 x값에 7을 곱하고 3을 더한 결과값인 y를 반환해주는 함수입니다.   
> n은 함수로부터 y값을 반환받아  52가 됩니다. 
>
> 함수를 호출하면서 함수에 넘겨주는 값을 `인수`라고 합니다. 위 코드에서는 7이 인수가 되겠죠.  
> `인수`를 전달 받는 x는 `매개 변수(parameter)`라고 부릅니다.  
> y와 같이 최종적으로 반환된 값을 `리턴 값`, `반환 값`이라고 부릅니다.   


#### **예제 3 : 두 수의 곱을 반환하는 함수 작성하기**

```python
## 예제_3_두_수의_곱을_반환하는_함수_작성하기. 
def product(a,b):
    return a*b

print(product(3,5))
```

## **실생활 프로젝트** 
**❓ 노래방 점수를 입력받아 점수 별로 멘트를 출력하는 함수를 만들어봅시다.**   

100점 : 환상적인 노래 실력, 가수 데뷔해도 되겠는데요?!  
90 ~ 99점 : 한곡 더 듣고 싶은 실력! 불러줄꺼죠?    
80 ~ 89점 : 탐나는 목소리를 가졌다고 전해라~    
70 ~ 79 : 한곡 더 놀아볼까?  
70점 미만 : 이 점수 어쩌면 좋지?  

#### **예제 4 : 노래방 멘트 출력하기**

```python
# 예제_4_노래방_멘트_출력하기
def singer(score):  # def 함수명(매개변수)
    score = int(score)
    if score == 100 :
        print("환상적인 노래 실력, 가수 데뷔해도 되겠는데요?!")
    elif score >= 90 and score < 100:
        print("한곡 더 듣고 싶은 실력! 불러줄꺼죠?")
    elif score >= 80 and score < 90:
        print("탐나는 목소리를 가졌다고 전해라~")
    elif score >= 70 and score < 80:
        print("한곡 더 놀아볼까?")
    else :
        print("이 점수 어쩌면 좋지?")
        
n = input("노래방 점수를 입력하세요!")
singer(n)
```

## **len함수**
* `len(리스트)` : 리스트의 길이를 반환합니다. 즉 리스트의 원소 개수를 반환합니다.   
* 예를 들어 a = [1,2,3,4]일 때, len(a)는 4를 반환합니다.  

```python
# 예제 5 : list를 매개 변수로 활용하기 (list 내에서 최소값을 리턴하기)
def minimum(numbers):
    min = numbers[0]  
        # numbers[0]을 일단 min값으로 저장하고, 이후 리스트의 원소들을 모두 순회하면서 min과 비교하고, min보다 작으면 해당 값을 min에 다시 저장해줍니다.
    for i in range(1,len(numbers)):  # i는 numbers의 원소들을 순회할 인덱스 넘버가 되어줄 것입니다. 따라서 1부터 len(numbers) 전까지만 순회해야 합니다. 
        if min > numbers[i]:
            min = numbers[i]  # min보다 해당 원소값이 더 작으면, 그 값을 min에 다시 저장해줍니다. 
            
    # 반복을 모두 끝내면 원소들을 모두 순회한 후이기 때문에, min에 저장된 값은 numbers의 최소값이 저장됩니다. 
    
    return min  # 결과인 min을 반환해줍니다. 

num = [70,20,30,25,27]
print(minimum(num))
    
```

## **지역변수와 전역변수** 
* 파이선에서 변수의 사용 범위는 함수를 기준으로 다르게 표현됩니다.  
* 변수를 함수 내부에서 선언하면, 해당 함수 안에서만 활용할 수 있고, 이를 `지역변수`라고 합니다.   
* 변수를 함수 밖에서 선언하면 코드의 모든 영역에서 활요할 수 있고, 이를 `전역변수`라고 합니다. 



#### **예제 6 : 지역변수와 전역변수**

```python
# 예제_6_지역변수와_전역변수
def print_codle():
    x = "codle"
    print(x)

print_codle()
# print(x) -> x는 전역변수가 아니기 때문에 실행하면 오류가 발생합니다.  
```
